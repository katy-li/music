```python
# Import CSV as DataFrame
df = pd.read_csv("Youtube_Hot100_data2.csv")
```

```python
# Drop all N/A values from the DataFrame
df.dropna()
```

```python
# View columns in the DataFrame
df.columns
```

```python
# Removes Scientific Notation from the DataFrame 
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
#Creating a Pivot Table 
artist_df_pivot= pd.pivot_table(df, index=['Song_artist', 'Song_title'], values=['likeCount', 'Song_duration_weeks', 'Song_peak', 'Song_rank', 'commentCount', 'dislikeCount','viewCount'])
```

```python
# Reset the Index to be on Song Artist and Song Title
youtube_df = artist_df_pivot.reset_index()
```

```python
# Check DataFrame
youtube_df
```

```python
youtube_df = youtube_df.fillna(0)
```

```python
youtube_df.dtypes
```

```python
cf.set_config_file(offline=False, world_readable=True, theme='pearl')

youtube_df.iplot(kind='bubble', x='Song_duration_weeks', y='viewCount', size='commentCount', text=['Song_artist','Song_title'],
             xTitle='Number of Weeks on Hot 100', yTitle='View Count',
             filename='yt')
```

```python
cf.set_config_file(offline=False, world_readable=True, theme='pearl')

youtube_df.iplot(kind='bubble', x='Song_duration_weeks', y='viewCount', size='dislikeCount', text=['Song_artist','Song_title'],
             xTitle='Number of Weeks on Hot 100', yTitle='View Count',
             filename='yt')
```

```python
cf.set_config_file(offline=False, world_readable=True, theme='pearl')

youtube_df.iplot(kind='bubble', x='Song_duration_weeks', y='viewCount', size='likeCount', text=['Song_artist','Song_title'],
             xTitle='Number of Weeks on Hot 100', yTitle='View Count',
             filename='yt')
```
