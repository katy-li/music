```python
uniqo = db.billy_boards_first_run.find()

weeks = []

for u in uniqo:
    week = u['currentChartWeek']
    if week not in weeks:
        weeks.append(week)
    
```

```python
total_weeks = len(weeks)
fw = total_weeks - 1
first_week_scrapped = weeks[fw]
most_recent_scrapped = weeks[0]

print("total_weeks: ",  total_weeks)
print("first_week: " + first_week_scrapped)
print("most_recent_week: " + most_recent_scrapped)
```
